```python
# get normalized occurrence matrix in a flat format and merge with totaltx
def get_daily_occ(x, y):
    daily_occcurence_input = np.array([], dtype=np.float32)
    for index, row in priced_bitcoin_input.iterrows():
        daily_occurrence_matrix_path_name = OCC_OUT_PATH + "\\dailyOccmatrices\\" + "occ" + str(row['year']) + '{:03}'.format(row['day']) + ".csv"
        daily_occurrence_matrix = pd.read_csv(daily_occurrence_matrix_path_name, sep=",", header=None).values
        # daily_occurrence_normalized_matrix = np.asarray(daily_occurrence_matrix).reshape(1, daily_occurrence_matrix.size)/row['totaltx']
        daily_occurrence_normalized_matrix = daily_occurrence_matrix[x - 1][y - 1]/row["totaltx"]
        # daily_occurrence_normalized_matrix = daily_occurrence_matrix[x - 1][y - 1]
        daily_occcurence_input = np.append(daily_occcurence_input, daily_occurrence_normalized_matrix)
    return daily_occcurence_input

def scale_prices(priced_bitcoin):

    price = priced_bitcoin['price'].values
    log_return_list = list()

    for index in range(len(price)):
        if index == 0:
            log_return_list.append(float(0))
        else:
            log_return_list.append(np.log(price[index]) - np.log(price[index - 1]))
    log_return_list = pd.DataFrame(log_return_list, columns=["log_return"])
    log_return_list.reset_index(drop=True, inplace=True)
    priced_bitcoin = pd.concat([priced_bitcoin, log_return_list], axis=1)

    return priced_bitcoin

def reg_m_rec(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    results = sm.RecursiveLS(y, X).fit()
    return results

def reg_m(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    results = sm.OLS(y, X).fit()
    return results

priced_bitcoin = pd.read_csv(PRICED_BITCOIN_FILE_PATH, sep=",")
# priced_bitcoin_label = scale_prices(priced_bitcoin)[['log_return']]
priced_bitcoin_label = priced_bitcoin[['price']]
priced_bitcoin_input = priced_bitcoin[['year', 'day', 'totaltx']]

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
totaltx_scaled = min_max_scaler.fit_transform(np.asarray(priced_bitcoin_input['totaltx']).reshape(-1,1)).reshape(-1)
daily_occcurence_input = get_daily_occ(1, 7)
daily_occcurence_input = min_max_scaler.fit_transform(daily_occcurence_input.reshape(-1,1)).reshape(-1)
input = np.concatenate((daily_occcurence_input, totaltx_scaled)).reshape(2,daily_occcurence_input.__len__())
```

```python
res = reg_m(priced_bitcoin_label.values.reshape(-1), input)
print(res.summary())
```

```python
res_rec = reg_m_rec(priced_bitcoin_label.values.reshape(-1), input)
print(res_rec.summary())
```
